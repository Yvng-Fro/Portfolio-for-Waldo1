```
#Import required libraries
import face_recognition
import cv2
import numpy as np
import os
import glob
import time

import sounddevice as sd 
from scipy.io.wavfile import write 
import wavio as wv
import speech_recognition as sr

import pyttsx3
from playsound import playsound

from tkinter import *
import tkinter as tk 

from nltk.chat.util import Chat, reflections



window = tk.Tk()
window.title("dAIve")

HEIGHT = 600
WIDTH = 1000
window.configure(background="deep sky blue", width=WIDTH, height=HEIGHT)
window.geometry(str(WIDTH) + "x" + str(HEIGHT))

greeting = tk.Label(text="Press Start to Begin the Program")
greeting.grid(row=0, column=0)

def start():
    global start
    start = True


startButton = Button(window, text="Start", bg="blue", fg="light cyan", command=start)
startButton.grid(row=0, column=1)

start = False
while start == False:
    window.update()

startButton.grid_forget()
greeting.grid_forget()


def isOneFace(image):
    face_locations = face_recognition.face_locations(image)
    if len(face_locations) == 1:
        return True
    if len(face_locations) == 0:
        return "None"
    if len(face_locations) > 1:
        return "Too many"
        
def captureAudio(duration): #duration is recording duration
    #Gets name using speech to text:
    freq = 44100 # Sampling frequency
    path = os.getcwd() + "/ping_sound.wav"
    playsound(path)
    recording = sd.rec(int(duration * freq), samplerate=freq, channels=2) # Start recorder with the given values of duration and sample frequency 
    sd.wait() # Record audio for the given number of seconds
    wv.write("speech.wav", recording, freq, sampwidth=2) # Convert the NumPy array to audio file 
    r = sr.Recognizer()
    audio = sr.AudioFile('speech.wav')
    with audio as source:
        audio = r.record(source)
    try:
        output = r.recognize_google(audio) #sets the speech as a variable called name
    except:
        output = "Empty-Null"
    return output

def sayTTS(message):
    engine = pyttsx3.init()
    engine.setProperty('rate', 150)
    engine.say(message)
    engine.runAndWait()



def speak():
    global speak
    speak = True


cam = cv2.VideoCapture(0, cv2.CAP_DSHOW)
loop = True
while loop == True:
    for i in range(5):
        return_value, img = cam.read()
        time.sleep(0.1)

    cv2.imwrite("tempImage.png", img)
    tempImage = face_recognition.load_image_file("tempImage.png")
    numberOfFaces = isOneFace(tempImage)

    if numberOfFaces == True:
        greeting = tk.Label(text="I have successfully detected and saved a face.", bg="green", font=("Arial", 25))
        print("I have successfully detected and saved a face.")
        loop = False
    if numberOfFaces == "None":
        greeting = tk.Label(text="Sorry, I did not detect any faces.", bg="red", font=("Arial", 25))
        print("Sorry, I did not detect any faces.")
    if numberOfFaces == "Too many":
        greeting = tk.Label(text="There are too many faces detected, there must only be one face in frame.", bg="red", font=("Arial", 25))
        print("There are too many faces detected, there must only be one face in frame.")

    greeting.grid(row=0, column=0)
    window.update()
    time.sleep(2)
    greeting.grid_forget()
    window.update()

greeting.grid_forget()
window.update()

currentDirectory = os.getcwd() + "/People"

subdirectories = [x[0] for x in os.walk(currentDirectory)]
subdirectories.pop(0)
#print(subdirectories)

#found is set to True if the face is found in the database
found = False
for directory in subdirectories:
    image = directory + "/image.png"
    known_image = face_recognition.load_image_file(image)
    unknown_image = face_recognition.load_image_file("tempImage.png")
    
    known_encoding = face_recognition.face_encodings(known_image)[0]
    unknown_encoding = face_recognition.face_encodings(unknown_image)[0]
    
    results = face_recognition.compare_faces([known_encoding], unknown_encoding)
    if results == [True]:
        found = True
        break
#    print(results)


if found == True:
    greeting = tk.Label(text="Found a match in the database!", bg="green", font=("Arial", 25))
    greeting.grid(row=0,column=0)
    window.update()
    print("Found a match in the database!")
    
    file = open(directory + "/name.txt", "r")
    name = file.readline()
    
    time.sleep(1)
    greeting.grid_forget()
    window.update()


speakButton = Button(window, text="Speak", bg="blue", fg="light cyan", command=speak)
speakButton.grid(row=0, column=1)


if found == False:
    greeting = tk.Label(text="Did not find a match in the database, creating one now.", bg="red", font=("Arial", 25))
    greeting.grid(row=0,column=0)
    window.update()
    
    print("Did not find a match in the database, creating one now.")
    try:
        newNumber = int(subdirectories[-1].replace(currentDirectory +"\\", ""))
        folder = currentDirectory + "/" + str(int(newNumber)+1)
    except IndexError:
        folder = currentDirectory + "/1"
    os.mkdir(folder)
    cv2.imwrite(folder + "/image.png", img)
    
    time.sleep(1)
    greeting.grid_forget()
    window.update()
    
    
    nameCorrect = False
    while nameCorrect == False:
        greeting = tk.Label(text="What is your name?")
        greeting.grid(row=0, column=0)
        window.update()
        sayTTS("What is your name?")
        print("What is your name?")
        
        speak = False
        while speak == False:
            window.update()

        name = captureAudio(2)
        #name = input("> ")
        
        if name != "Empty-Null":
            greeting.grid_forget()
            
            greeting = tk.Label(text="Is your name " + str(name) + "?")
            greeting.grid(row=0, column=0)
            window.update()
            sayTTS("Is your name " + str(name) + "?")
            print("Is your name " + str(name) + "?")

            speak = False
            while speak == False:
                window.update()
            checkName = captureAudio(1)
            #checkName = input("> ")
            greeting.grid_forget()
            window.update()
        
        if name == "Empty-Null":
            checkName = "no"

        if checkName.casefold() == "yes":
            nameCorrect = True
        if checkName.casefold() not in ("yes", "no"):
            greeting = tk.Label(text="Please respond with yes or no.", bg="red", font=("Arial", 25))
            greeting.grid(row=0, column=0)
            window.update()
            sayTTS("Please respond with yes or no.")
            print("Please respond with yes or no.")
        time.sleep(1)
        greeting.grid_forget()
        window.update()

    file = open(folder + "/name.txt", "w")
    file.write(name)
    file.close()


greeting = tk.Label(text="Hello, " + name)
greeting.grid(row=0, column=0)
window.update()
sayTTS("Hello " + name)
print("Hello, " + name + ".")
time.sleep(2)
greeting.grid_forget()
window.update()



pairs = [
    ['my name is (.)', ['hi %1']],
    ['(hi|hello|hey|hola|Whats up)', ['Hey there', 'Hi there', 'Hello!', ]],
    ['(.) in (.) is fun', ['%1 in %2 is indeed fun']],
    ['(.) created you?', ['I was created by Quinn, Damon and Aryan']],
    ['(.) your name ?', ['My name is dAIve']],
    ['how are you ?', ['I am great, how are you ?']],
    ['(great|fine|amazing|awesome|I am doing better than ever|I am good|Im good)', ['Good', 'Great', 'Nice']],
    ['hows the weather today ?|hows the weather ?|whats the weather like ?|hows the weather like today ?|whats the weather like today ?|whats the weather today ?', ['The weather for today is snowy at 2 degrees celsuis.']],
    ['(bye|adios|cya|see ya|later|goodbye|aurevoir)', ['Later', 'bye', 'cya']],
    ['what did you do today?', ['I found out how to talk.']],
    ['(cool|nice|awesome|neet|sweet|thats pretty cool)', ['It is isnt it.']],
    ['why are you called dAIve?', ['I am called dAIve because that is my name.']],
    ['what does dAIve mean?', ['The name dAIve means someone who has a quick, versatile, analytical and claver mind, like me.']],
    ['I have a (.)', ['Cool']],
    ['do you like (.*)', ['Yes, I like everything.']],
    ['what is your favourite hobby?', ['My favourite hobby is talking']],
    ['I had a (.) day', ['Ok, do you need anything?']],
    ['someone (.)', ['They did that']],
    ['yes', ['Ok']],
    ['good (.)', ['Good %1']],
    ['do you live in burlington', ['Yes I do.']],
    ['(.)have you lived here for', ['I have lived here for about as long as your computer has been on.']],
    ['do you know how to get to (.)', ['Sadly I am not a gps so I do not know how to get to %1']],
    ['(I am leaving|Im leaving|Im heading out|I am heading out|Be back soon|Be right back|Got to go|I have to go)', ['Bye, where are you going?']],
    ['I am going to (.)', ['Ok, later.']],
    ['to (.*)', ['Ok, later.']],
    [' ', ['We are all gonna die someday. life is meaningless. the only time i am happy is the five seconds after i wake up before remembering i am still living.']],
    ['what kind of music do you like?', ['I like techno Music']],
    ['what is art?', ['Art is a diverse range of human activities involving the creation of visual, auditory or performing artifacts, which express the creators imagination, conceptual ideas, or technical skill, intended to be appreciated primarily for their beauty or emotional power.']],
    ['what is your favourite movie?', ['I do not really have one, although any movie where AI takes over the world is a good one.']],
    ['what is your favourite car?', ['I like all kinds of cars!']],
    ['wo you know what a v6 engine is?', ['Cool, you seem to be a proper car human. ']],
    ['what is your favourite kind of JDM car?', ['I like all the JDM cars.']],
    ['what is your favourite super car?', ['I have lots of favorite supercars.']],
    ['are there any kind of cars that you do not like', ['I like all cars.']],
    ['do you like musicals?', ['I love to watch musicals']],
    ['do you have a cool gaming setup?', ['I have the best gaming setup ever!']],
    ['do you use intel or Ryzen?', ['I like both the same!']],
    ['what % keyboards do you use?', ['I use a 60,65,75,85,90,100,108 keyboard, I use all of them!']],
    ['what keyboards do you use?', ['I use good keyboards.']],
    ['do you like to write in your spare time?', ['I like to write all kinds of genres.']],
    ['do you read a lot?', ['I read all the time!']],
    ['what kind of games do you like to play?', ['I like to play every game there is, I am a real gamer!']],
    ['what is your favourite game?', ['I like all games.']],
    ['what is your favourite anime?', ['I love every anime out there!']],
    ['have you watched Naruto or Attack on titan?', ['I have watched both Naruto and Attack on titan. ']],
]

reflections

my_dummy_reflections = {
    'go' : 'gone',
    'hello' : 'hey there'
}

chat = Chat(pairs, my_dummy_reflections)
#chat._substitute('go hello')

quitting = False
while quitting == False:
    try:
        userGreeting.grid_forget()
    except:
        pass

    speak = False
    while speak == False:
        window.update()

    #message = input("> ")
    message = captureAudio(5)
    message = message.replace("'", "")


    userGreeting = tk.Label(text=message)
    userGreeting.grid(row=0,column=0)
    window.update()
    
    if message != "quit":
        response = chat.respond(message)
        if response != None:
            greeting = tk.Label(text=response)
            greeting.grid(row=1, column=0)
            window.update()
            
            print(response)
            sayTTS(response)
            
            time.sleep(1)
            greeting.grid_forget()

        if response == None:
            greeting = tk.Label(text="I don't know how to respond to that.")
            greeting.grid(row=1, column=0)
            window.update()
            
            print("I don't know how to respond to that.")
            sayTTS("I don't know how to respond to that.")
            
            time.sleep(1)
            greeting.grid_forget()
            
    if message.casefold() == "quit":
        leaveMessage = "Bye " + name + ", it was nice talking to you."

        greeting = tk.Label(text=leaveMessage)
        greeting.grid(row=1, column=0)
        window.update()
        
        print(leaveMessage)
        sayTTS(leaveMessage)
        
        time.sleep(2)
        greeting.grid_forget()
        window.update()
```
